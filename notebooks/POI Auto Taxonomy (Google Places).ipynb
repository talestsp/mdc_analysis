# Analysing Both Sides Occurrences

```python
both_sides = []

for categ in pd.Series(types_list).drop_duplicates():
    lr_categ = left_right_term(categ, d["types"])
    lr = term_placement_analisis(lr_categ, show=False)

    for related_categ, row in lr["both"].iterrows():
        rate = max(row["left"] / row["right"], row["right"] / row["left"])
        
        both_sides.append({"categ": categ,
                           "related": related_categ,
                           "left_freq": row["left"], 
                           "right_freq": row["right"],
                           "rate": rate})
        
both_sides = pd.DataFrame(both_sides)[["categ", "related", "rate", "left_freq", "right_freq"]]
```

```python
def remove_words_from_list(lista, words_to_remove=["establishment", "point_of_interest"]):
    lista = copy.deepcopy(lista)
    for word_to_remove in words_to_remove:
        if word_to_remove in lista:
            lista.remove(word_to_remove)
    return lista
```

```python
relations_all_freq = relations_all_df.groupby("parent")["category"].value_counts().to_frame().rename({'category': 'freq'}, axis=1).reset_index().sort_values("freq", ascending=False)
relations_all_freq.head(12)
```

```python
relations_visited_freq = relations_visited_df.groupby("parent")["category"].value_counts().to_frame().rename({'category': 'freq'}, axis=1).reset_index().sort_values("freq", ascending=False)
relations_visited_freq.head(12)
```

```python
from src.utils.type_hierarchy_analysis import parent, children

def occurrences_in_all_pois(term, pois_types=d["types"], round_n=3):
    occurrences = pois_types.apply(lambda types : term in types).value_counts()
    try:
        freq = occurrences.loc[True]
    except KeyError:
        freq = 0
        
    prop = freq / len(pois_types)
    
    if round_n:
        prop = round(prop, round_n)
    
    return {"freq": freq, "prop": prop}

def occurrences_in_visited_pois(term, user_tags_sequence=users_tags_sequence, round_n=3):
    freq = 0
    n_pois = 0
    
    for user_id in user_tags_sequence.keys():
        n_pois += len(user_tags_sequence[user_id])
        for types in user_tags_sequence[user_id]:
            if term in types:
                freq += 1
    
    prop = freq / n_pois
    
    if round_n:
        prop = round(prop, round_n)
    
    return {"freq": freq, "prop": prop}
```

```python
def len_categs(categories):
    total = 0
    for key in categories.keys():
        total += len(categories[key])
    return total

def is_categ(categ, categories):
    for key in categories.keys():
        if categ in categories[key]:
            return True
    return False
```

# Comparison Visited POIs vs All POIs

```python
all_vs_visited = []

for a_type in types_list.drop_duplicates().reset_index(drop=True):
    if a_type == "point_of_interest" or a_type == "establishment" or a_type == "premise":
        continue
    all_pois = occurrences_in_all_pois(a_type)
    visited_pois = occurrences_in_visited_pois(a_type)
    row = {"type": a_type, 
           "freq_all": all_pois["freq"], "prop_all": all_pois["prop"],
           "freq_visited": visited_pois["freq"], "prop_visited": visited_pois["prop"]}
    
    all_vs_visited.append(row)
    
all_vs_visited = pd.DataFrame(all_vs_visited).sort_values("freq_visited", ascending=False)

corr_spearman_head = all_vs_visited["prop_all"].head(30).corr(all_vs_visited["prop_visited"].head(30), method="spearman")
corr_spearman = all_vs_visited["prop_all"].corr(all_vs_visited["prop_visited"], method="spearman")

corr_pearson_head = all_vs_visited["prop_all"].head(30).corr(all_vs_visited["prop_visited"].head(30), method="pearson")
corr_pearson = all_vs_visited["prop_all"].corr(all_vs_visited["prop_visited"], method="pearson")

print("Comparing frequencies of visited pois against all pois")
print("Spearman top 30 items: {}".format(corr_spearman_head))
print("Pearson  top 30 items: {}".format(corr_pearson_head))
print()
print("Spearman all items   : {}".format(corr_spearman))
print("Pearson  all items   : {}".format(corr_pearson))

all_vs_visited = all_vs_visited.reset_index(drop=True).reset_index(drop=False).set_index("type")
all_vs_visited.head(30)
```
